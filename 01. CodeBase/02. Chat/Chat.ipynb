# 4. Response Generation Unit

<br>

## 4.1. Importing custom Packages

```python
# importing custom packages
from Chatpkg import ResponseHandler
from Chatpkg import TextProcessor
```

<br>

## 4.2. Response Generation

```python
def text_generate(customer_message, CacheDB):
    # generating prediction
    predicted_intent = utils.get_prediction(customer_message)

    if predicted_intent == 'not_sure':
        return {'reply': 'Sorry, I can not understand.', 'CacheDB': CacheDB, 'predicted_intent': predicted_intent}



    # ---------------------------------------
    #         Response : Rule Based         |
    # ---------------------------------------
    elif response_type[predicted_intent] == 'RuleBased':
        res = ResponseHandler.RuleBased(predicted_intent, CacheDB, customer_message)
        
        return {'reply': res['reply'], 'CacheDB': res['CacheDB'], 'predicted_intent': predicted_intent}


    
    # ---------------------------------------
    #         Response : Task Based         |
    # ---------------------------------------
    elif response_type[predicted_intent] == 'TaskBased':
        res = ResponseHandler.TaskBased(predicted_intent, CacheDB, customer_message)
        
        return {'reply': res['reply'], 'CacheDB': res['CacheDB'], 'predicted_intent': predicted_intent}



    # ---------------------------------------
    #         Response : Something Wrong    |
    # ---------------------------------------
    else:
        return {'reply': 'Sorry, something went wrong.', 'CacheDB': CacheDB, 'predicted_intent': predicted_intent}
```

# 5. Chat

```python
# initiating the one time intents
CacheDB['one_time_intents'] = []

# printing the first message
print(f'Qiana : Hello !\n')

# appending the one time intent for first message
CacheDB['one_time_intents'].append('Greetings')

# continue chat till customer exits
while True:
    # taking customer input
    customer = input('Customer : ')

    if customer.lower() == 'q':
        break

    # generating the response
    res = text_generate(customer, CacheDB)

    reply = res['reply']
    CacheDB = res['CacheDB']

    # dispaying the response
    print(f'Qiana : {reply}')

    print('\n')
```
