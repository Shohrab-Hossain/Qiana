> Validating the Column Names

```python
# reanaming the column name in small letter
df.columns = list( map(lambda x: x.lower(), df.columns) )
```

```python
df.head()
```

<br>

> Validating the Item Names

```python
def modify(item):
    splitted = item.split()

    to_remove = ['-'] # list of character to remove
    for keyword in to_remove: # looping to check for each character
        if keyword in splitted:
            splitted.remove(keyword)
    new_item = ' '.join(splitted)
    new_item = new_item.strip()
    return new_item

# removing special character from item name
df[item_name] = df[item_name].apply(modify)

# reanaming the item names in small letter
df[item_name] = df[item_name].apply(lambda x: x.lower())

df.head()
```

<br>

> Validating the Menu

```python
# reanaming the menu in small letter
df[menu] = df[menu].apply(lambda x: x.lower())
```

```python
df[menu].value_counts()
```

<br>
<br>

## Creating Food Bag-Of-Word

```python
item_names = list(df[item_name])

for item in item_names:
    # food_bow.extend( list( filter(lambda x: x.isalnum(), item.split() ) ) )
    food_bow.extend(item.split())

food_bow = list( set(food_bow) )
food_bow.sort()
```

## Writing the generated data locally

<br>

> Writting the BOW in JSON format

```python
FoodBowReady = {'Food_BOW' : food_bow}

filename = 'Food_BOW.json'
filepath = f'{rootDir}/Generated Data/{filename}'

with open(filepath, 'w') as f:
    json.dump(FoodBowReady, f)
```
