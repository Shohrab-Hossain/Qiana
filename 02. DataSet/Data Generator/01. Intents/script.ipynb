## Defining Data Structure

## Reading Intents List

```python
fileName = 'intentsList.json'
filepath = f'{rootDir}/01. Intents/{fileName}'

with open(filepath) as f:
  data = json.load(f)

  intentsList = data['intents']
```

```python
number_of_intents = len(intentsList)
print(f'Total Number of intents : {number_of_intents}')
```

<br>
<br>

## Functions to Validate Intent


```python
def validateIntent(intent):
  # accessing the intent file name
  global filename

  # Checking whether any key is missing in the intent
  key_to_validate = ['intent', 'response_type', 'texts']

  for _key in key_to_validate:
    if _key not in intent.keys():
      raise Exception(f"'{_key}' key is missing in the '{filename}' file.")

  return True
```

<br>
<br>

## Reading Intent from local storage and validating

```python
for index, intentName in enumerate(intentsList):
  filename = intentName + '.json'
  filepath = f'{rootDir}/01. Intents/Raw Single Intent/{filename}'

  try:
    with open(filepath) as f:
        intent = json.load(f)

    if validateIntent(intent):
      intents.append(intent)

  except FileNotFoundError:
    print(f"File Missing Error : '{filename}' file is missing.")

  else:
    print(f"> Success : {index+1} of {number_of_intents} intents read complete.")

print("\n\n")
print("All intents have been read successfully.")
```

<br>
<br>

## Writing the generated intents

```python
intentsReady = {'intents' : intents}

filename = 'intents.json'
filepath = f'{rootDir}/Generated Data/{filename}'

with open(filepath, 'w') as f:
    json.dump(intentsReady, f)
```
